```python
import pandas as pd
import numpy as np
def getMeanVariables(hdf5_file,features):
    pdf_full=pd.read_hdf(hdf5_file, key='df')
    return pdf_full[features].mean()

def getMeanErr(hdf5_file,features):
    pdf_full=pd.read_hdf(hdf5_file, key='df')
    return pdf_full[features].std()/np.sqrt(len(pdf_full))

hdfs = [f'/lustre/isaac/scratch/tmengel/jet-background-subtraction/background-subtraction/ml-subtract/results/{ds}_results.h5' for ds in datasets ]
hdfs_auau = [f'/lustre/isaac/scratch/tmengel/jet-background-subtraction/background-subtraction/ml-subtract/results/{ds}_results.h5' for ds in datasets[:3]]
hdfs_pbpb = [f'/lustre/isaac/scratch/tmengel/jet-background-subtraction/background-subtraction/ml-subtract/results/{ds}_results.h5' for ds in datasets[3:]]

mean_rho_auau = np.array([getMeanVariables(hdfs_auau[i],features =['median_pt_over_npart']) for i in range(3)])
mean_rho_pbpb = np.array([getMeanVariables(hdfs_pbpb[i],features =['median_pt_over_npart']) for i in range(3)])
mean_rho_error_auau = np.array([getMeanErr(hdfs_auau[i],features =['median_pt_over_npart']) for i in range(3)])
mean_rho_error_pbpb = np.array([getMeanErr(hdfs_pbpb[i],features =['median_pt_over_npart']) for i in range(3)])

mean_npart_auau = np.array([getMeanVariables(hdfs_auau[i],features =['jet_nparts_pythia']) for i in range(3)])
mean_npart_pbpb = np.array([getMeanVariables(hdfs_pbpb[i],features =['jet_nparts_pythia']) for i in range(3)])
mean_npart_error_auau = np.array([getMeanErr(hdfs_auau[i],features =['jet_nparts_pythia']) for i in range(3)])
mean_npart_error_pbpb = np.array([getMeanErr(hdfs_pbpb[i],features =['jet_nparts_pythia']) for i in range(3)])
```
